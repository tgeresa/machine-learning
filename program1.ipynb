```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#to ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(r"C:\Users\HP\Desktop\MACHINE LEARNING\heartdataset.csv")
```

```python
# Display basic information about the dataset
print(data.info())
print(data.head())
```

```python
# Define numeric columns
numeric_columns = ['Age', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak', 'HeartDisease']

# Basic statistics
print(data[numeric_columns].describe())
```

```python
for col in numeric_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(data[col], kde=True, bins=20, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Plot boxplots
for col in numeric_columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=data[col], color='orange')
    plt.title(f'Box Plot of {col}')
    plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = data[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairplot for numeric columns
sns.pairplot(data[numeric_columns], diag_kind='kde')
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```
